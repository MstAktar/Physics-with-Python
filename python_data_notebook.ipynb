```python
# common libraries
import numpy as np #numeric calculation
import pandas as pd #data manipulations
import matplotlib.pyplot as plt #graphics
```

```python
#read the data
df = pd.read_csv('RawData.csv')
df.head()
df.info()
```

```python
df.head()
```

```python
#To Access only single column
df['Time (s)']
```

```python
#To Access only single column
#df['Time (s)']

#Let's make simply graphics
plt.figure(figsize=(15,15))

plt.subplot(3,1,1)
plt.plot(df['Time (s)'], df['Linear Acceleration x (m/s^2)'])
plt.grid()
plt.title('Acceleration x')
plt.xlabel('Time')
plt.ylabel('a_x')
plt.axis([0,10,-30,30])

plt.subplot(3,1,2)
plt.plot(df['Time (s)'], df['Linear Acceleration y (m/s^2)'])
plt.grid()
plt.title('Acceleration y')
plt.xlabel('Time')
plt.ylabel('a_y')
plt.axis([0,10,-30,30])

plt.subplot(3,1,3)
plt.plot(df['Time (s)'], df['Linear Acceleration z (m/s^2)'])
plt.grid()
plt.title('Acceleration z')
plt.xlabel('Time')
plt.ylabel('a_z')
plt.axis([0,10,-30,30])
plt.show()
```
